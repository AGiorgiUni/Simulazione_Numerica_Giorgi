```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

n= np.loadtxt("Esercizio_1/dati.txt",usecols=(0), delimiter=' ',   unpack='true')

c=np1.loadtxt("Esercizio_1/dati.txt",usecols=(1), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio_1/dati.txt", usecols=(2), delimiter=' ', unpack='true')

n2= np.loadtxt("Esercizio_1/dati2.txt",usecols=(0), delimiter=' ',   unpack='true')

c2=np1.loadtxt("Esercizio_1/dati2.txt",usecols=(1), delimiter=' ', unpack='true')

error2 = np2.loadtxt("Esercizio_1/dati2.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(n,c,yerr=error)
plt.errorbar(n2,c2,yerr=error2)
plt.xlabel('Throws')
plt.ylabel('Call')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

n= np.loadtxt("Esercizio_1/dati.txt",usecols=(0), delimiter=' ',   unpack='true')

c=np1.loadtxt("Esercizio_1/dati.txt",usecols=(3), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio_1/dati.txt", usecols=(4), delimiter=' ', unpack='true')

n2= np.loadtxt("Esercizio_1/dati2.txt",usecols=(0), delimiter=' ',   unpack='true')

c2=np1.loadtxt("Esercizio_1/dati2.txt",usecols=(3), delimiter=' ', unpack='true')

error2 = np2.loadtxt("Esercizio_1/dati2.txt", usecols=(4), delimiter=' ', unpack='true')

plt.errorbar(n,c,yerr=error)
plt.errorbar(n2,c2,yerr=error2)
plt.xlabel('Throws')
plt.ylabel('Put')
plt.grid(True)
plt.show()
```
