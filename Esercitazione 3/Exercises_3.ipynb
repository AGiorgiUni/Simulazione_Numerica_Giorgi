# Exercise 1


I estimate European put and call option prices sampling a geometric Brownian motion. The first dataset is computed directly using only one gaussian number, while the other is computed step by step with more estimation of the price at different times. 

I simulate 10000 estimation for every method and than I compute  statistical uncertainties using mean std, dividing in blocks of 100. Final results are compared to Black-Scholes analytic solutions: call=14.975790778311286 and put=5.4595325819072364

## Call

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2


f  = plt.figure(figsize=(15,4))
axes1 = plt.subplot(1,2,1)

plt.title("Direct estimation")

n= np.loadtxt("Esercizio_1/dati.txt",usecols=(0), delimiter=' ',   unpack='true')

c=np1.loadtxt("Esercizio_1/dati.txt",usecols=(1), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio_1/dati.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(n,c,yerr=error)
plt.xlabel('Throws')
plt.ylabel('Call')
plt.grid(True)


axes2 = plt.subplot(1,2,2)
plt.title("Discrete estimation")
n2= np.loadtxt("Esercizio_1/dati2.txt",usecols=(0), delimiter=' ',   unpack='true')

c2=np1.loadtxt("Esercizio_1/dati2.txt",usecols=(1), delimiter=' ', unpack='true')

error2 = np2.loadtxt("Esercizio_1/dati2.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(n2,c2,yerr=error2)
plt.xlabel('Throws')
plt.ylabel('Call')
plt.grid(True)

plt.show()

print("Direct final value = "+str(c[-1])+" +/ "+str(error[-1]))
print("Final value by steps = "+str(c2[-1])+" +/ "+str(error2[-1]))
```

The two final estimations are consistent each other with Black-Scholes analytic solution. Note that discrete simulation is more computationally difficult, but guarantees a faster convergence to the mean value.
