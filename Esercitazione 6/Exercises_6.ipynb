# Metropolis


In the first part, Metropolis sampling is used and values are obtained with block (of 10000 steps) sampling. First question is equilibration: I plot the results for a large number of steps in ordere to check when they reach a stable value.
The first simulation of the Ising model has been done at T=0.5 with natural units. One can check "by eye" that this method needs nearly 60 blocks for equilibration.

```python
x_en= np.loadtxt("ISING_1D/output.ene.0",usecols=(0), delimiter=' ',   unpack='true')
f_en=np.loadtxt("ISING_1D/output.ene.0",usecols=(2), delimiter=' ', unpack='true')
error_en= np.loadtxt("ISING_1D/output.ene.0", usecols=(3), delimiter=' ', unpack='true')
x_heat= np.loadtxt("ISING_1D/output.Heat.0",usecols=(0), delimiter=' ',   unpack='true')
f_heat=np1.loadtxt("ISING_1D/output.Heat.0",usecols=(2), delimiter=' ', unpack='true')
error_heat = np.loadtxt("ISING_1D/output.Heat.0", usecols=(3), delimiter=' ', unpack='true')
x_mag= np.loadtxt("ISING_1D/output.Mag.0",usecols=(0), delimiter=' ',   unpack='true')
f_mag=np.loadtxt("ISING_1D/output.Mag.0",usecols=(2), delimiter=' ', unpack='true')
error_mag = np.loadtxt("ISING_1D/output.Mag.0", usecols=(3), delimiter=' ', unpack='true')
x_chi= np.loadtxt("ISING_1D/output.Chi.0",usecols=(0), delimiter=' ',   unpack='true')
f_chi=np1.loadtxt("ISING_1D/output.Chi.0",usecols=(2), delimiter=' ', unpack='true')
error_chi = np.loadtxt("ISING_1D/output.Chi.0", usecols=(3), delimiter=' ', unpack='true')

plt.figure(figsize=(30, 30))
fig, ax = plt.subplots(2, 2, figsize=(26, 20))

ax[0,0].set_title("Energy")
ax[0,0].plot(x_en, f_en, color="blue", linewidth=2.5)
ax[0,0].errorbar(x_en,f_en,yerr=error_en,linewidth=2)
ax[0,0].set_xlabel('Block number')
ax[0,0].set_ylabel('<Potential energy>')
ax[0,0].grid(True)

ax[0,1].set_title("Heat capacity")
ax[0,1].plot(x_heat, f_heat,color="blue",linewidth=2.5)
ax[0,1].errorbar(x_heat,f_heat,yerr=error_heat, linewidth=2)
ax[0,1].set_xlabel('Block number')
ax[0,1].set_ylabel('<Heat capacity>')
ax[0,1].grid(True)

ax[1,0].set_title("Magnatization")
ax[1,0].plot(x_mag, f_mag,color="blue",linewidth=2.5)
ax[1,0].errorbar(x_mag,f_mag,yerr=error_mag, linewidth=2)
ax[1,0].set_xlabel('Block number')
ax[1,0].set_ylabel('<Magnatization>')
ax[1,0].grid(True)

ax[1,1].set_title("Susceptibility")
ax[1,1].plot(x_chi, f_chi,color="blue",linewidth=2.5)
ax[1,1].errorbar(x_chi,f_chi,yerr=error_chi, )
ax[1,1].set_xlabel('Block number')
ax[1,1].set_ylabel('<Susceptibility>')
ax[1,1].grid(True)


plt.show()
```

After 60 epochs, all the measures reaches a quite stable value. In this plot, magnetization is calculated with h=0.0 but it has a more regular behaviour using h=0.02, so the same number of epochs for equilibration is adopted.
