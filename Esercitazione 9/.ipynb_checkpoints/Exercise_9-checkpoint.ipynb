# Travelling Salesman

My project is realized using a matrix of 100 configurations as number of columns and 34 indices for every city in rows: my idea is not to change the first culumn which reports the best sequence for every step and to modify the other ones. 

The first matrix is prepered just copying 100 times the sequence of cities as they are genereted, then two types of mutation are proposed with probability $p_1$ and $p_2$: the first consists in swapping two consecutive cities with $50\%$ of probability, the second one selects at least 3 cities and inverts their order. First reorganiztion is executed and it calculates cost function ($\sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|$) for every column and the best sequence is reported in every columns with even indices. Then, crossover is proposed with probbility $p_3$ for every column: this one consists in selecting a random index for row and then reorder the successive elements as they appear in the next column. Finally, a new reorganiztion is executed and this process is repeted $n_{steps}$ times. All the mutations don't affect the starting city (number 0) and cost function includes the distance between the last city and the fisrt one. After reotganization this matrix appears like:


| c_0 | c_1 | c_2 | c_3 |
| --- | --- | --- | --- |
| 0 | 0 | 0 | 0 |
| 2 | 4 | 2 | 5 |
| 3 | 6 | 3 | 3 |
| 22 | 15 | 22 | 22 |
| 10 | 12 | 10 | 15 |
| 7 | 30 | 7 | 29 |
| 17 | 22 | 17 | 34 |
| 5 | 1 | 5 |4 |


## First test

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x= np.loadtxt("es1/output.dist_in_circle.dat",usecols=(0), delimiter=' ',   unpack='true')
f=np.loadtxt("es1/output.dist_in_circle.dat",usecols=(1), delimiter=' ', unpack='true')

x1= np.loadtxt("es1/output.cord_in_circle.dat",usecols=(0), delimiter=' ',   unpack='true')
f1=np.loadtxt("es1/output.cord_in_circle.dat",usecols=(1), delimiter=' ', unpack='true')

fig,axs = plt.subplots(1, 2,figsize=(15,15))


axs[0].plot(x1,f1,marker='o')
axs[0].set_title('Cartesian Coordinates', fontsize=15)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].grid(True)
axs[0].set_box_aspect(1)

axs[1].plot(x,f,marker='o')
axs[1].set_title('Cost function', fontsize=15)
axs[1].set_xlabel('Steps')
axs[1].set_ylabel('Cost function')
axs[1].grid(True)
axs[1].set_box_aspect(1)

plt.show()
```

The first test places 34 cities on a circumference and the best travel is found as the first plot shows. Different probabilities for mutations and crossover have been tested:

| p_1 | p_2 | p_3 | steps for convergence |
| --- | --- | --- |--- |

| 20\% | 20\%  | 50\% |>1500 |
| 5\% | 20\%  | 50\% |1400 |
