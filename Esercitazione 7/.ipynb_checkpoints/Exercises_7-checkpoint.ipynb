# Study of autocorrelation

These datasets are computed by plotting 500000 steps of liquid, solid and gas simulation using Metropolis methods: I use 500000 blocks of 1 step. No equilibration for T and for the system occurs.

### Liquid

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


x= np.loadtxt("Liquid/MDNVE_MCNVT/output_epot.dat",usecols=(0), delimiter=' ',   unpack='true')

f=np.loadtxt("Liquid/MDNVE_MCNVT/output_epot.dat",usecols=(1), delimiter=' ', unpack='true')

error = np.loadtxt("Liquid/MDNVE_MCNVT/output_epot.dat", usecols=(3), delimiter=' ', unpack='true')

print("Total number of data =", f.size)
#plt.errorbar(x,f,yerr=error)
plt.xlabel('Lags')
plt.ylabel('Correlation rate')
plt.grid(True)

plt.acorr(f, maxlags = 500)
plt.show()

n_blocks=50
print("Number of steps for each block =", f.size/n_blocks)
final = np.array_split(f,n_blocks)
x=np.array(0)
for i,m in enumerate(final):
    x=np.append(x,np.average(final[i]))
x=np.delete(x, 0)
plt.acorr(x,maxlags = n_blocks-1)
plt.xlabel('Lags')
plt.ylabel('Correlation rate')
plt.grid(True)
plt.show()

n_blocks=10
print("Number of steps for each block =", f.size/n_blocks)
final = np.array_split(f,n_blocks)
x=np.array(0)
for i,m in enumerate(final):
    x=np.append(x,np.average(final[i]))
x=np.delete(x, 0)
plt.acorr(x,maxlags = n_blocks-1)
plt.xlabel('Lags')
plt.ylabel('Correlation rate')
plt.grid(True)
plt.show()
```

# Study of error in data blocking

Now, check how error, calulated as mean std, change increasing number of steps (L) for each block:

```python
import numpy as np
import matplotlib.pyplot as plt
f=np.loadtxt("Solid/MDNVE_MCNVT/output_epot.dat",usecols=(1), delimiter=' ', unpack='true')
M=f.size
x=np.array(0)
L_array=np.array(0)

fig,axs = plt.subplots(1, 3,figsize=(15, 5))


L=10
while L <= 5000:
    L_array=np.append(L_array,L)
    final = np.array_split(f,M/L)
    x=np.append(x,np.std(final[1])/(final[1].size-1))
    L=L+10
x=np.delete(x, 0)
L_array=np.delete(L_array, 0)

axs[0].plot(L_array,x)
axs[0].set_title("Error in solid",fontsize=16)
axs[0].set_xlabel("L")
axs[0].set_ylabel("Mean std")


f=np.loadtxt("Liquid/MDNVE_MCNVT/output_epot.dat",usecols=(1), delimiter=' ', unpack='true')
M=f.size
x=np.array(0)
L_array=np.array(0)
L=10
while L <= 5000:
    L_array=np.append(L_array,L)
    final = np.array_split(f,M/L)
    x=np.append(x,np.std(final[1])/(final[1].size-1))
    L=L+10
x=np.delete(x, 0)
L_array=np.delete(L_array, 0)
axs[1].plot(L_array,x)
axs[1].set_title("Error in liquid",fontsize=16)
axs[1].set_xlabel("L")



f=np.loadtxt("gas/MDNVE_MCNVT/output_epot.dat",usecols=(1), delimiter=' ', unpack='true')
M=f.size
x=np.array(0)
L_array=np.array(0)
L=10
while L <= 5000:
    L_array=np.append(L_array,L)
    final = np.array_split(f,M/L)
    x=np.append(x,np.std(final[1])/(final[1].size-1))
    L=L+10
x=np.delete(x, 0)
L_array=np.delete(L_array, 0)
axs[2].plot(L_array,x)
axs[2].set_title("Error in gas",fontsize=16)
axs[2].set_xlabel("L")

plt.show()
```

Mean std progressivly decreases with the number of steps for each block: every simulation confirms the central limit theorem.

The further simulations are realized in the same starting conditions (except for starting T) of exercise 4, but now I use NVE and 20 blocks of 10000 steps for MC without equilibration (not necessary beacuse T is stable). I include also measure for radial distrbution g(r) which is compared for the two methods.

## Liquid phase

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x= np.loadtxt("Liquid/MDNVE_MCNVT/output_epot1.dat",usecols=(0), delimiter=' ',   unpack='true')
f=np.loadtxt("Liquid/MDNVE_MCNVT/output_epot1.dat",usecols=(2), delimiter=' ', unpack='true')
error = np.loadtxt("Liquid/MDNVE_MCNVT/output_epot1.dat", usecols=(3), delimiter=' ', unpack='true')

x1= np.loadtxt("Liquid/MDNVE_MCNVT/output_pres1.dat",usecols=(0), delimiter=' ',   unpack='true')
f1=np.loadtxt("Liquid/MDNVE_MCNVT/output_pres1.dat",usecols=(2), delimiter=' ', unpack='true')
error1 = np.loadtxt("Liquid/MDNVE_MCNVT/output_pres1.dat", usecols=(3), delimiter=' ', unpack='true')

fig,axs = plt.subplots(1, 2,figsize=(15, 5))

axs[0].errorbar(x,f,yerr=error)
axs[0].set_title('Potential energy')
axs[0].set_xlabel('Block number')
axs[0].set_ylabel('<Potential energy>')
axs[0].grid(True)

axs[1].errorbar(x1,f1,yerr=error1)
axs[1].set_title('Pressure')
axs[1].set_xlabel('Block number')
axs[1].set_ylabel('<Pressure>')
axs[1].grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
y=np.array(0)
error=np.array(0)
i=0
while i<100:
    y= np.append(y,np.average(np.loadtxt("Liquid/MDNVE_MCNVT/output_gdr1.dat",usecols=(i), delimiter=' ',   unpack='true')))
    error=np.append(error,np.std(np.loadtxt("Liquid/MDNVE_MCNVT/output_gdr1.dat",usecols=(i), delimiter=' ',   unpack='true'))/19)
    i=i+1
y=np.delete(y, 0)
error=np.delete(error, 0)

y1=np.array(0)
error1=np.array(0)
i=0
while i<100:
    y1= np.append(y1,np.average(np.loadtxt("Liquid/MDNVE_MCNVT/output_gdr.dat",usecols=(i), delimiter=' ',   unpack='true')))
    error1=np.append(error1,np.std(np.loadtxt("Liquid/MDNVE_MCNVT/output_gdr.dat",usecols=(i), delimiter=' ',   unpack='true'))/19)
    i=i+1
y1=np.delete(y1, 0)
error1=np.delete(error1, 0)

x=np.arange(100)

fig,axs = plt.subplots(1, 2,figsize=(15, 5))
axs[0].errorbar(x,y,yerr=error)
axs[0].set_title("g(r) normalized in box dimension with MC")
axs[0].set_xlabel('r [box/2]')
axs[0].set_ylabel('g(r)')
axs[0].grid(True)

axs[1].errorbar(x,y1,yerr=error1)
axs[1].set_title("g(r) normalized in box dimension with NVE")
axs[1].set_xlabel('r [box/2]')
axs[1].set_ylabel('g(r)')
axs[1].grid(True)
plt.show()
```
