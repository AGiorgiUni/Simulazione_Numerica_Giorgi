## Exercise 11.2

Different tests have been realized to reach the final model: mse as loss is straightforward beacuse the aim is to minimize the distance beetween predicted ad training points, but choice of optimizer depends on the model and gsd is used because it seems to work well with relu. Relu as activation function for all the layers is adopted because it gives the best results: different combination have been checked, but adopting a sigmoid or other limited function produces a vary limited output and other activation functions similar to relu (as elu) produce worse performance. Only the last layers uses elu because it smoothes the predicted function. 

This problem recquires more parameters to be set and more images than the past one because of function complexity, so 5 layers with different neurons are used to optimize the  parameters number using 1500 training images.
