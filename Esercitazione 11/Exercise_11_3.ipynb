```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import math
```

```python
n_training=500
n_validation=50

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-3/2, 3/2, n_training)
x_valid = np.random.uniform(-3/2, 3/2, n_validation)
y_train = np.random.uniform(-3/2, 3/2, n_training)
y_valid = np.random.uniform(-3/2, 3/2, n_validation)
x_valid.sort()
y_valid.sort()
z_target=np.array(math.sin(x_valid[0]**2+y_valid[0]**2))
i=1
while i<(n_validation):
  z_target = np.append(z_target,math.sin(x_valid[i]**2+y_valid[i]**2) )
  i=i+1


sigma = 1.0 
z_train=np.array(math.sin(x_train[0]**2+y_train[0]**2))
z_valid=np.array(math.sin(x_valid[0]**2+y_valid[0]**2))
i=1
while i<n_training:
      z_train = np.append(z_train,np.random.normal(math.sin(x_train[i]**2+y_train[i]**2), sigma) )
      i=i+1

i=1
while i<n_validation:
  z_valid= np.append(z_valid,np.random.normal(math.sin(x_valid[i]**2+y_valid[i]**2), sigma))
  i=i+1


array=np.append(x_train,y_train)
array=np.reshape(array,(n_training,2))
train=tf.convert_to_tensor(array) 
array=np.append(x_valid,y_valid)
array=np.reshape(array,(n_validation,2))
valid=tf.convert_to_tensor(array) 
```

```python
# compose the NN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(2,), activation='relu'))
#model.add(Dense(3, input_shape=(2,), activation='sigmoid'))
model.add(Dense(3, input_shape=(2,), activation='sigmoid'))
model.add(Dense(3, input_shape=(2,),activation='sigmoid'))
model.add(Dense(1, input_shape=(2,),activation='tanh'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
# get a summary of our composed model
model.summary()
```

```python


# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=train, y=z_train,
          batch_size=32, epochs=10,
          shuffle=True, # a good idea is to shuffle input before at each epoch. Mischia input a ogni epoca
          validation_data=(valid,z_valid))
```

```python
# evaluate model
score = model.evaluate(valid,z_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
#plt.plot()
plt.savefig("/content/drive/MyDrive/Colab Notebooks/Es11.2/es_11.3.jpg")
```
