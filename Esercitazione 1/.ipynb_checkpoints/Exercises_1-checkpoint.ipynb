# Exercise 1.1

The purpose of this exercise is to test a uniform number generator. So numbers beetween 0 and 1 have been generated and tested. Firstly, progressive mean value of them has been studied:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

x= np.loadtxt("Esercizio1/dati.txt",usecols=(0), delimiter=' ',   unpack='true')

f=np1.loadtxt("Esercizio1/dati.txt",usecols=(1), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio1/dati.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('Throws')
plt.ylabel('<x>')
plt.grid(True)
plt.show()
```

As expected from central limit theorem, generating ramndomly numbers in [0,1] the mean value converges to 0.5 and the mean square error decreases with higher throws. Then, convergence of $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ is studied:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

x= np.loadtxt("Esercizio1/dati2.txt",usecols=(0), delimiter=' ',   unpack='true')

f=np1.loadtxt("Esercizio1/dati2.txt",usecols=(1), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio1/dati2.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('Throws')
plt.ylabel('<sigma_square> - 1/12')
plt.grid(True)
plt.show()
```

Like the previous test, the integral converges to 0 and error decreases with higher value of throws. 
At the end, I checked if numbers are uniformly spread in [0,1]: I diveded the range in 100 sub-ranges and then I counted how many numbers of the 1000 generated were located in the sub intervals. The expected value is 100, so I checked the obtained value with a chi square test

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

x= np.loadtxt("Esercizio1/dati3.txt",usecols=(0), delimiter=' ',   unpack='true')

f=np1.loadtxt("Esercizio1/dati3.txt",usecols=(1), delimiter=' ', unpack='true')

plt.errorbar(x,f)
plt.xlabel('throws')
plt.ylabel('<chi_square>')
plt.grid(True)
plt.show()
```

It shows that chi value for every throw is approximatly 100. This implies that $\chi_f=\sum \chi_i$ with $\chi_i=1$ and so, inverting the chi-formula used, it means that numbers in 100 subintervals generating uniformly 10000 numbers are $n_i=M/N+\sqrt{N/M} \sim 100$, as expected.

Cauchy-Lorentz distribution is caracterized by a higher peak and lower tails than the gaussian one. The last histogram shows that peak is higher than the last one of the firsr distribution and standard daviation is 10 times lower, so we can conclude that this distribution is similar to a Lorentzian one.

# Exercize 1.3

In this exercise I tried to measure $\pi$ with Buffon's experiment and random uniform generator of numbers. I generated the initial position of a needle and its direction with an angle in $[- \pi /2, \pi/2]$ from the orthogonal to the straight line. I set L=3 and D=7.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

x= np.loadtxt("Esercizio3/pi.txt",usecols=(0), delimiter=' ',   unpack='true')

f=np1.loadtxt("Esercizio3/pi.txt",usecols=(1), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio3/pi.txt", usecols=(2), delimiter=' ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('Throws')
plt.ylabel('<pi>')
plt.grid(True)
plt.show()
```

From the graphics, there's the evidence that there is convergence to a lower value of $\pi$ ($\sim 3.09$), but with higher number of steps, measured value continue to increase to 3.14.  This problem is not related to the distance D beacuse the mean value of $\pi$ don't change too much with D.
