# Molecular Dynamics

These simulations are realized using 50 epochs of 2000 steps and every phase has been equilibrated: the three models start from a different temperature than the requested one untill they reach it after some steps. Infact, this algorithm introduces a scale factor for velocities which perturbs the system, kinetik energy and so temperature. After these equilibrations steps real simulation starts and requested measure are calculated with data blocking. Equilibration steps and starting temperatures are reported:

| Phase | starting T | Requested T | steps of eq |
| --- | --- | --- | --- |
| Liquid | 1.98 | 1.1 | 150 |
| Gas | 0.95 | 1.2 | 1500 |
| Solid | 1.55 | 0.8 | 220 |

# Liquid Phase

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy as np1
import numpy as np2

x= np.loadtxt("Esercizio_1/Liquid/output_epot.dat",usecols=(0), delimiter=' ',   unpack='true')

f=np1.loadtxt("Esercizio_1/Liquid/output_epot.dat",usecols=(2), delimiter=' ', unpack='true')

error = np2.loadtxt("Esercizio_1/Liquid/output_epot.dat", usecols=(3), delimiter=' ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('Block number')
plt.ylabel('<Potential energy>')
plt.grid(True)
plt.show()
```
